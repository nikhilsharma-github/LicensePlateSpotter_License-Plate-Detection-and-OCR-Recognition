```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import cv2
```

Importing our labels.cv and printing the result

```python
df=pd.read_csv('labels.csv')
df.head()
```

Importing xml.etree library for xml navigation and handling 

```python
import xml.etree.ElementTree as xet
```

```python
#printing the first filepath in the images folder 
filename=df['filepath'][0]
filename
```

```python
#Creating a function, which will obtain the file_path and file_name as text
def getFilename(filename):
    filename_image=xet.parse(filename).getroot().find('filename').text
    filepath_image=os.path.join('./images',filename_image)
    return filepath_image
```

```python
#printing the filename
getFilename(filename)
```

```python
df['filepath'].apply(getFilename)
```

```python
#converting the filepath and filenames in a list and printing it.
image_path=list(df['filepath'].apply(getFilename))
image_path
```

#### Data Preprocessing 

```python
#Installing tensorflow for the Data Preprocessing

# pip install tensorflow==2.11.0
```

```python
#Importing important libraries for the Data Preprocessing.
#sklearn:  python library to implement machine learning models and statistical modelling.
#Through scikit-learn, we can implement various machine learning models for 
#regression, classification, clustering, and statistical tools for analyzing these models
#tensorflow: It is used to manage everything regarding machine learning models
#keras: A high level Neural Network library used to run Deep Learning Models, 
#a sublibrary for Tensorflow

from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
#This is used to obtain all the values in an array
labels = df.iloc[:,1:].values
```

```python
#Creating a Generalised function here.
#We are creating 2 lists, data and output here.
#For every image, obtain the image in a variable, get height, width and depth of image
# do preprocessing on image, which is resizing it to a particular coordinate
# converting image coordinates to array, and doing normalization of the values.

#after this, doing normalization of labels which is xmin, xmax, ymin and ymax.
# At last, appending these values to data list and adding the result to output list.
data=[]
output=[]
for ind in range(len(image_path)):
    image=image_path[ind]
    img_arr = cv2.imread(image)
    h,w,d = img_arr.shape
    #preprocessing
    load_image=load_img(image,target_size=(224,224))
    load_image_arr=img_to_array(load_image)
    norm_load_image_arr=load_image_arr/255.0 #normalization
    #normalization to labels
    xmin,xmax,ymin,ymax=labels[ind]
    nxmin,nxmax=xmin/w,xmax/w
    nymin,nymax=ymin/h,ymax/h
    label_norm=(nxmin,nxmax,nymin,nymax) #normalized output
    data.append(norm_load_image_arr)
    output.append(label_norm)
```

```python
output
```

```python
#splitting the data into training and testing sets.
#Also, converting it to float type
X=np.array(data,dtype=np.float32)
y=np.array(output,dtype=np.float32)
```

```python
#printing the number of data with the dimensions and number of outputs.
X.shape,y.shape
```

```python
#splitting the data into training and testing sets. Taking train size as 80%
#rest goes in testing (20%), and printing below.

x_train,x_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=0)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## Deep Learning Model

```python
#importing libraries and models.
#Tensorflow.keras.applications models like MobileNetv2, InceptionV3, InceptionResNetV2.
#These are powerful and good object detection models.

#Importing tensorflow layers as Dense, Flatten, Dropout, Input
#Importing the tensorflow model library, and tensorflow as tf

from tensorflow.keras.applications import MobileNetV2, InceptionV3, InceptionResNetV2
from tensorflow.keras.layers import Dense, Dropout, Flatten, Input
from tensorflow.keras.models import Model
import tensorflow as tf
```

```python
#Library used to store huge amount of numerical data
#pip install h5py==3.8.0
```

```python
#We are gonna use inceptionResNetV2 Learning model here.

#weight as 'imagenet', shape=224,224,3. 
inception_resnet=InceptionResNetV2(weights="imagenet",include_top=False,
                                  input_tensor=Input(shape=(224,224,3)))
#Marking as False because we don't need to change the weight of images.
inception_resnet.trainable=False

#main convolutional neural network model.
headmodel=inception_resnet.output
#Flattening the headmodel.
headmodel=Flatten()(headmodel)

#taking the headmodel

#Number of neurons for the neural network layer 1.
headmodel=Dense(500,activation="relu")(headmodel)

#Number of neurons for the neural network layer 2.
headmodel=Dense(250,activation="relu")(headmodel)

# Final output of neural network should be 4, sigmoid gives output between 0-1
headmodel=Dense(4,activation='sigmoid')(headmodel)

#Setting all the configurations of our model below
model=Model(inputs=inception_resnet.input,outputs=headmodel)
```

```python
#Compiling our model

#loss is 'MeanSquarEerror', optimizer here is Adam Optimzer.
# learning rate should be less as possible(0.001) i.e 1e-4
model.compile(loss='mse',optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4))

#Printing the model summary below.
model.summary()
```

## Model Training of InceptionResNetV2

```python
#Importing tensorflow callback from TensorFlow

from tensorflow.keras.callbacks import TensorBoard
```

```python
#Importing the log directory of Tensorflow 'object_detection'

tfb = TensorBoard('object_detection')
```

```python
# printing the history of the model training below.
# x is xtrain. y is ytrain.
# batch size is the collection of images being trained in one go.
# epochs is the iterations of the training of the batch images.
# validation data is being xtest and ytest here.
# for callbacks, we are using tfb imported above.

#As per the speed of our cpu, we need to take the appropriate batch_size and epochs
history=model.fit(x=x_train,y=y_train,batch_size=10,epochs=50,validation_data=(x_test,y_test),callbacks=[tfb])
```

```python
#After completing the above training process.
# We are again performing the same operation.
# taking the same batch size, making epochs from 51 to 100 (50).
# this is being performed to reduce the losses and improve the accuray of training

history=model.fit(x=x_train,y=y_train,batch_size=10,epochs=100,validation_data=(x_test,y_test),
                  callbacks=[tfb],initial_epoch=51)
```

```python
# Till now, the model is being stable after the total compilation and training.
# losses being reduced to the lowest possible number and now we are saving our model.

#Saving the model in the folder 'models' as 'object_detection.h5'

model.save('./models/object_detection.h5')
```
