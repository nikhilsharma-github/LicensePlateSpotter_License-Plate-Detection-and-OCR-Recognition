Importing All required Libraries <br>
Pandas: Used to work with datasets and do analysis on it. <br>
xmletree: used to parse and navigate an xml file.<br>
glob: Used to extract the file path.

```python
import pandas as pd
import xml.etree.ElementTree as xet
```

```python
from glob import glob
```

Here, we are extracting all the file path using glob function

```python
path = glob('./images/*.xml')
path
```

```python

labels_dict=dict(filepath=[],xmin=[],xmax=[],ymin=[],ymax=[])
for filename in path:

    #parsing the file
    info = xet.parse(filename)
    #getting the root of the file
    root = info.getroot()
    #obtaining the file which is object as nature
    member_object = root.find('object')
    #getting the boundbox coordinate in labels_info from the object
    labels_info = member_object.find('bndbox')
    #storing the coordinates of boundbox in variables below.
    xmin = int(labels_info.find('xmin').text)
    xmax = int(labels_info.find('xmax').text)
    ymin = int(labels_info.find('ymin').text)
    ymax = int(labels_info.find('ymax').text)
    #print(xmin,xmax,ymin,ymax)
    
    #In the dictionary, we are saving the filename, the xmin, ymin,
    # xmax and ymax coordinate
    labels_dict['filepath'].append(filename)
    labels_dict['xmin'].append(xmin)
    labels_dict['xmax'].append(xmax)
    labels_dict['ymin'].append(ymin)
    labels_dict['ymax'].append(ymax)
```

```python
#storing the file dictionary as dataframe in df variable and printing below.
df =  pd.DataFrame(labels_dict)
df
```

```python
#printing the dictionary to double check the result.
labels_dict
```

```python
#Now, here we are using the to_csv function which will take 
# our dataframe df variable and give us the csv version of the result
# the file will be saved to the current directory.
df.to_csv('labels.csv',index=False)
```
